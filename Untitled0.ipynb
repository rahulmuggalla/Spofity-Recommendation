```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Load the Spotify dataset
spotify_df = pd.read_csv('/content/drive/MyDrive/ML Flask Projects/Spotify/spotify.csv')

spotify_df
```

```python
# User input for song and number of recommendations
song_input = input("Enter a song: ")
num_recommendations = int(input("Enter the number of recommendations: "))

# Filter the dataset to get the song's data
song_data = spotify_df[spotify_df['name'].str.contains(song_input, case=False)]

# Extract the relevant features for similarity calculation
features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness',
            'speechiness', 'tempo', 'valence']

# Calculate the similarity between songs using cosine similarity
vectorizer = CountVectorizer().fit_transform(song_data['artists'])
artist_similarity = cosine_similarity(vectorizer)

# Get the index of the input song
input_song_index = song_data.index[0]

# Get the most similar songs based on artist similarity
similar_songs_indices = artist_similarity[input_song_index].argsort()[::-1][:num_recommendations + 1]
similar_songs_indices = similar_songs_indices[similar_songs_indices != input_song_index]

# Get the recommended songs and artists
recommended_songs = spotify_df.loc[similar_songs_indices, 'name'].tolist()
recommended_artists = spotify_df.loc[similar_songs_indices, 'artists'].tolist()

# Print the recommended songs and artists
print(f"Recommended songs based on '{song_input}':")
for i in range(len(recommended_songs)):
    print(f"{i + 1}. Song: {recommended_songs[i]}, Artist: {recommended_artists[i]}")
```
